# Monte Carlo for Calculating Pi

Let's calculate $\pi$! We're going to generate random pairs of numbers over a unit square and then calculate the number of pairs inside a circle inscribed in the unit square. Since $A=\pi r^2$ and $r=0.5$ we can estimate $\pi \approx 4A$ where $A$ is the number of random points inside the circle divided by the total number of points. We can tell is a random point $(x,y)$ is inside a circle by seeing if its $L_2$ norm is less than or equal to 1. That is, if $x^2 + y^2 \leq 1$ then $(x,y)$ is inside the circle.

```python
import random
import matplotlib.pyplot as plt
import math

N = 1000
x_s = [-1+2*float(x/N) for x in range(0,N+1)]
```

```python
y1=[]
y2=[]
for i in range(len(x_s)):
    y1.append(math.sqrt(1-x_s[i]**2))
for i in range(len(y1)):
    y2.append(-y1[i])
```

```python
plt.plot(x_s,y1)
plt.plot(x_s,y2)
plt.show()
```

```python
def monte(N_pts):
    count = 0
    for i in range(0,N_pts):
        x = 2.0*random.random()-1
        y = 2.0*random.random()-1
        if(x**2+y**2 <= 1):
            count+=1
    return 4.0*float(count/N_pts)
```

```python
trials = [10**n for n in range(1,7)]
pi_appr = []
for i in range(len(trials)):
    pi_appr.append(monte(trials[i]))
```

```python
plt.plot(range(1,7), pi_appr)
assympi = [math.pi for i in range(1,7)]
plt.plot(range(1,7), assympi)
plt.show()
```
