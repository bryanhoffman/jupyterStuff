```python
#following along with course located here: https://github.com/numerical-mooc/numerical-mooc/blob/master/lessons/01_phugoid/01_02_Phugoid_Oscillation.ipynb

import numpy
from matplotlib import pyplot
%matplotlib inline
```

```python
T = 100.0
dt = 0.002
N = int(T/dt)+1
t=numpy.linspace(0.0,T,N)
g=9.81
```

```python
#ICs
z0 = 100.0
b0 = -20.0
zt = 100.0

u = numpy.array([z0,b0])

z = numpy.zeros(N)
z[0] = z0
```

```python
for n in range(1,N):
        u = u + dt*numpy.array([u[1],g*(1-u[0]/zt)])
        z[n] = u[0]
```

```python
#first time using pyplot here
pyplot.figure(figsize=(10,4))
pyplot.ylim(min(z)-10,max(z)+10)
pyplot.tick_params(axis='both', labelsize=24)
pyplot.xlabel('t',fontsize=14)
pyplot.ylabel('z',fontsize=14)
pyplot.plot(t,z,'k-')
```

```python
from math import sin, cos, log, ceil
import numpy
from matplotlib import rcParams
rcParams['font.family'] = 'serif'
rcParams['font.size'] = 16
```

```python
v_t = 30.0
C_D = 1/40
C_L = 1

v_0 = v_t
theta0 = 0
x0 = 0
y0 = 1000
```

```python
def f(u):
    v = u[0]
    theta = u[1]
    x = u[2]
    y = u[3]
    return numpy.array([-g*sin(theta)-(C_D*g*v**2)/(C_L*v_t**2),-g*cos(theta)/v + g*v/v_t**2,v*cos(theta),v*sin(theta)])
```

```python
def euler_step(u,f,dt):
    return u + dt*f(u)
```

```python
T = 100.0
dt = 0.01
N = int(T/dt)+1

u = numpy.empty((N,4))
u[0] = numpy.array([v_0,theta0,x0,y0])

for n in range(N-1):
    u[n+1] = euler_step(u[n], f, dt)
```

```python
pyplot.figure(figsize=(8,6))
pyplot.grid(True)
pyplot.xlabel(r'x', fontsize=18)
pyplot.ylabel(r'y', fontsize=18)
pyplot.title('Glider trajectory, flight time = %.2f' % T, fontsize=18)
pyplot.plot(u[:,2],u[:,3],'k-',lw=2)
```
