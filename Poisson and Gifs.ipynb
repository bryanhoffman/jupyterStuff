```python
import numpy as np
import matplotlib.pyplot as plt

from matplotlib import animation, rc
from IPython.display import HTML, Image
```

```python
rc('animation', html='html5')
```

```python
fig, ax = plt.subplots()

ax.set_xlim(( 0, 24))
ax.set_ylim((0, 6))

line, = ax.plot([],[],lw=1)
```

```python
def init():
    line.set_data([], [])
    return (line,)
```

```python
import random
y=np.zeros(180)
y_s=[]
y_s.append(y)
```

```python
def sample():
    poisson = np.zeros(180)
    for i in range(1,180):
        if (random.randrange(0, 10) == 0):
            poisson[i] = 1
    
    return poisson
```

```python
for i in range(1,200):
    y_s.append(y_s[i-1]+sample())
```

```python
def animate(i):
    x = np.linspace(0, 24, 180)
    y=y_s[i]
    line.set_data(x, y)
    return (line,)
```

```python
anim = animation.FuncAnimation(fig, animate, init_func=init,
                               frames=30, interval=150, blit=True)
```
